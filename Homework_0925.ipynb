```python
#숙제 2

import math

class Dice:
    def rollin_dice(self):
        return math.floor((random() * 6) + 1) 
        
player1 = Dice() 
player1_num = player1.rollin_dice() 

player2 = Dice() 
player2_num = player2.rollin_dice() 

print('player1_num: {0}, player2_num: {1}'.format(player1_num, player2_num))
      
if player1_num > player2_num:
      print('player1 승리')
elif player1_num < player2_num:
      print('player2 승리')
else:
      print('무승부!')
```

```python
#숙제 3

#예외처리 미구현

import math

num_of_player = int(input('플레이할 인원 수: '))

secret_store = False

class Dice:
    def rollin_dice(self):
        return math.floor((random() * 6) + 1)

#class Player:
    #progress = 0
    
players_progress = list()
players_money = list()
players_first_chk = list()
players_second_chk = list()
players_cursed_status = list()

loop_on = False

for i in list(range(num_of_player)): 
    players_progress.append(0)
    players_money.append(0)
    players_first_chk.append(0)
    players_second_chk.append(0)
    players_cursed_status.append(0)

d = Dice()

while 1:
    num_list_of_this_turn = list()
    
    for i in list(range(num_of_player)):
        
        cnt = 0
        for j in range(len(players_cursed_status)): #연속 굴리기는 돌지만 i차례는 연속굴리기가 아닌 차례를 스킵용
            if 0 <= players_cursed_status[j]:
                cnt += 1
            if cnt == num_of_player:
                loop_on = False
            else:
                loop_on = True
        
        if players_cursed_status[i] < 0: 
            players_cursed_status[i] += 1
            print(i, '번째 플레이어가 연속으로 굴립니다. 연속 굴리기', -players_cursed_status[i], '번 남았습니다.')
            
            loop_on = True
            
            if players_cursed_status[i] == 0:
                loop_on = False
                    
        if 0 <= players_cursed_status[i] and loop_on == True: #연속 굴리기
            continue
        
        if players_cursed_status[i] > 0:
            players_cursed_status[i] -= 1
            print(i, '번째 플레이어가 정지 상태이므로 건너뜁니다. 정지 상태', players_cursed_status[i], '번 남았습니다.')
            print()
            continue
        
        ran_num = d.rollin_dice()
        print(i, '번째 플레이어의 주사위 값:', ran_num)
        
        num_list_of_this_turn.append(ran_num)
        
        if ran_num == 2:
            conduct_or_no = input('주사위의 특수 능력을 활용하시겠습니까?(y, n): ')
            
            if conduct_or_no == 'y':
                print('모든 플레이어의 자금 -50')
                
                for j in list(range(num_of_player)):
                    players_money[j] -= 50
                    print(j, '번째 플레이어의 현재 자금:', players_money[j])
                    
        elif ran_num == 4:
            players_money[i] = math.floor(players_money[i] * 0.95)
            print(i, '번째 플레이어의 자금이 5퍼센트 차감되어', players_money[i], '이 되었습니다.')
            
        elif ran_num == 6:
            print('모든 플레이어의 자금 +30')
            
            for j in list(range(num_of_player)):
                    players_money[j] += 30
                    print(j, '번째 플레이어의 현재 자금:', players_money[j])
        
        players_progress[i] += ran_num
        
        print('현재', i,'번째 플레이어의 진행도:', players_progress[i])
        
        if 40 <= players_progress[i] and players_first_chk[i] == 0:
            players_first_chk[i] = 1
            print(i,'번째 플레이어의 한 바퀴 완주 완료! 자금 100 추가!')
            players_money[i] += 100
            print('현재', i, '번째 플레이어의 자금:', players_money[i])
            print()
            
        if 80 <= players_progress[i] and players_second_chk[i] == 0:
            players_second_chk[i] = 1
            print(i,'번째 플레이어의 두 바퀴 완주 완료! 자금 100 추가!')
            players_money[i] += 100
            print('현재', i, '번째 플레이어의 자금:', players_money[i])
            print()
        
        print()
        cnt = 0
            
        for j in range(len(num_list_of_this_turn)):
            
            if num_list_of_this_turn[j] % 2 == 0:
                cnt += 1
                
        if cnt == num_of_player:
            print('비밀상점 오픈!')
            secret_store = True
            
        if(secret_store):
            print('비밀 상점 물품: 레드 티켓 - 100만원, 옐로 티켓 - 50만원, 카오스 티켓 - 10만원, 뽑기 티켓 - 10만원, 연속 굴리기 - 20만원')
            print() 
                  
            for k in list(range(num_of_player)):
                print(k, '번째 플레이어의 아이템 선택 차례입니다.')
                product_index = int(input('선택할 아이템 번호(1 ~ 5, 취소: 0)): '))
                
                if product_index == 1:
                    
                    if 100 <= players_money[k]:
                        players_money[k] -= 100
                        print(k,'번째 플레이어가 레드 티켓을 구매하셨습니다. 현재 자금:', players_money[k])

                        targer_player = int(input('아이템을 대상으로 할 플레이어 번호: '))
                        print(targer_player, '번째 플레이어의 주사위 턴이 +5회 정지됩니다!')

                        players_cursed_status[targer_player] += 5
                        print()
                    
                    else:
                        print('자금이 부족합니다!')
                        k -= 1 #되돌아가서 재선택이 안됨
                    
                elif product_index == 2:
                    
                    if 50 <= players_money[k]:
                        players_money[k] -= 50
                        print(k,'번째 플레이어가 옐로 티켓을 구매하셨습니다. 현재 자금:', players_money[k])

                        targer_player = int(input('아이템을 대상으로 할 플레이어 번호: '))
                        print(targer_player, '번째 플레이어의 주사위 턴이 +2회 정지됩니다!')

                        players_cursed_status[targer_player] += 2
                        print()
                        
                    else:
                        print('자금이 부족합니다!')
                        k -= 1
                    
                elif product_index == 3:
                    
                    if 10 <= players_money[k]:  
                        players_money[k] -= 10
                        print(k,'번째 플레이어가 카오스 티켓을 구매하셨습니다. 현재 자금:', players_money[k])
                        
                        random_target = math.floor((random() * num_of_player))
                        print(random_target, '번째 플레이어의 주사위 턴이 +2회 정지됩니다!')
                        
                        players_cursed_status[random_target] += 2
                        print()
                    
                    else:
                        print('자금이 부족합니다!')
                        k -= 1
                        
                elif product_index == 4:
                    
                    if 10 <= players_money[k]:  
                        players_money[k] -= 10
                        print(k,'번째 플레이어가 뽑기 티켓을 구매하셨습니다. 현재 자금:', players_money[k])
                        
                        random_box = math.floor(random() * 7) + 1
                        add_money = [1, 2, 5, 10, 20, 50, 100]
                        
                        if random_box == 1:
                            players_money[k] += 1
                        elif random_box == 2:
                            players_money[k] += 2
                        elif random_box == 3:
                            players_money[k] += 5
                        elif random_box == 4:
                            players_money[k] += 10
                        elif random_box == 5:
                            players_money[k] += 20
                        elif random_box == 6:
                            players_money[k] += 50
                        elif random_box == 7:
                            players_money[k] += 100
                            
                        print('당첨된 뽑기 티켓 번호:', random_box)
                        print(add_money[random_box - 1], '원이 추가되어 현재', k, '번째 플레이어의 금액: ', players_money[k])
                        print()
                        
                    else:
                        print('자금이 부족합니다!')
                        k -= 1
                        
                elif product_index == 5:
                    
                    if 20 <= players_money[k]:
                        players_money[k] -= 20
                        print(k,'번째 플레이어가 뽑기 티켓을 구매하셨습니다.')
                        
                        random_additional_turns = math.floor(random() * 4) + 1
                        print(k, '번째 플레이어가', random_additional_turns, '번 연속 굴리기에 당첨되었습니다!')
                        
                        players_cursed_status[k] -= random_additional_turns
                        print()
                        
                    else:
                        print('자금이 부족합니다!')
                        k -= 1
                        
                else:
                    k -= 1 #이게 안먹히네
                    print(k, '상점 물품 범위 내의 값을 입력해주세요!')
                 
            secret_store = False
            print()
    
    if 120 <= players_progress[i]:
        break
        
money_of_winner = 0
num_of_winner = -1
for i in list(range(num_of_player)):

    num = players_money[i]
    
    if money_of_winner < num:
        money_of_winner = num
        num_of_winner = i
    
print('우승자는', money_of_winner, '원을 보유한', num_of_winner, '번째 플레이어!!!')
    
    
    
    
```
