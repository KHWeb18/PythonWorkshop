```python
from random import *

class Dice:
    def __init__(self, player):
        self.player = player
        
    def run_dice(self):
        self.dice = randint(1, 6)
        print('{0} - 주사위 눈금은 {1}'.format(self.player, self.dice))
        
    def who_win(self, versus):
        if (self.dice > versus.dice):
            print("{0} 승리".format(self.player))
        elif (self.dice < versus.dice):
            print("{0} 승리".format(versus.player))
        else:
            print("비김 {0} vs {1}".format(self.dice, versus.dice))
            
    def get_dice(self):
        return self.dice

player1 = Dice('플레이어1')
player1.run_dice()

player2 = Dice('플레이어2')
player2.run_dice()

player1.who_win(player2)
player2.who_win(player1)
```

```python
from random import * 
import math

class Game:
    def __init__(self):
        self.secret_store = False
        self.game_end = False
        self.use_chaos_ticket = False
        self.player_arr = []
        self.dice_sum = 0
        self.loop_player_no = -1
        self.loop_on = False
        
    def open_secret_store(self):
        print("비밀 상점이 열렸으니 플레이어들은 물품을 구매해주세요!!!")
        # TODO: 숙제
        # 비밀 상점 기능은 실제 루프 돌면서 주사위를 돌리는 플레이어들의 행동을 제한해야함
        # 혹은 연속 돌리기 기능의 경우 다른 플레이어들은 정지 상태로 자신만 돌아야함
        
        for i in range(len(self.player_arr)):
            
            isWrongIndex = True
            
            while isWrongIndex:
                choosen_product = int(input('레드 티켓 - 100만원. 옐로 티켓 - 50만원, 카오스 티켓 - 10만원, 뽑기 티켓- 10만원, 연속 굴리기 - 20만원, 구매 스킵(0번)'))

                if choosen_product == 1:
                    
                    if 1000000 <= self.player_arr[i].money:
                        self.player_arr[i].money -= 1000000
                    
                        for j in range(len(self.player_arr)):
                            print(j + 1, '번', self.player_arr[j].player)

                        effect_target = int(input('레드 티켓의 목표 플레이어 선택: '))

                        print('플레이어', self.player_arr[effect_target - 1].player, '가 레드 티켓의 타깃으로 지정되었습니다.')            
                        self.player_arr[effect_target - 1].tied_times += 5

                        isWrongIndex = False
                    
                    else:
                        print('돈이 부족합니다. 다른 상품을 선택해주세요!')
                        isWrongIndex = True

                elif choosen_product == 2:

                    if 500000 <= self.player_arr[i].money:
                        self.player_arr[i].money -= 500000
                        
                        for j in range(len(self.player_arr)):
                            print(j + 1, '번', self.player_arr[j].player)

                        effect_target = int(input('옐로 티켓의 목표 플레이어 선택: '))

                        print('플레이어', self.player_arr[effect_target - 1].player, '가 옐로 티켓의 타깃으로 지정되었습니다.')            
                        self.player_arr[effect_target - 1].tied_times += 2

                        isWrongIndex = False
                    
                    else:
                        print('돈이 부족합니다. 다른 상품을 선택해주세요!')
                        isWrongIndex = True

                elif choosen_product == 3:
                    
                    if 100000 <= self.player_arr[i].money:
                        self.player_arr[i].money -= 100000

                        ran_target = randint(0, self.player - 1)

                        print('플레이어', self.player_arr[ran_target].player, '가 카오스 티켓의 타깃으로 지정되었습니다.')            
                        self.player_arr[ran_target].tied_times += 2

                        isWrongIndex = False
                        
                    else:
                        print('돈이 부족합니다. 다른 상품을 선택해주세요!')
                        isWrongIndex = True

                elif choosen_product == 4:
                    
                    if 100000 <= self.player_arr[i].money:
                        self.player_arr[i].money -= 100000

                        ran_num = randint(1, 7)

                        if ran_num == 1:
                            ran_money = 10000
                            self.player_arr[i].money += ran_money
                        elif ran_num == 2:
                            ran_money = 20000
                            self.player_arr[i].money += ran_money
                        elif ran_num == 3:
                            ran_money = 50000
                            self.player_arr[i].money += ran_money
                        elif ran_num == 4:
                            ran_money = 100000
                            self.player_arr[i].money += ran_money
                        elif ran_num == 5:
                            ran_money = 200000
                            self.player_arr[i].money += ran_money
                        elif ran_num == 6:
                            ran_money = 500000
                            self.player_arr[i].money += ran_money
                        elif ran_num == 7:
                            ran_money = 1000000
                            self.player_arr[i].money += ran_money

                        print('플레이어', self.player_arr[i].player, '가', ran_money, '를 뽑기 티켓으로 얻었습니다.')

                        isWrongIndex = False
                    
                    else:
                        print('돈이 부족합니다. 다른 상품을 선택해주세요!')
                        isWrongIndex = True

                elif choosen_product == 5:
                    
                    if 200000 <= self.player_arr[i].money:
                        self.player_arr[i].money -= 200000

                        random_additional_turns = randint(1, 4)
                        print('연속 굴리기를 구매해', random_additional_turns, '회 연속 굴립니다.')

                        self.player_arr[i].add_times += (random_additional_turns + 1)

                        isWrongIndex = False
                    
                    else:
                        print('돈이 부족합니다. 다른 상품을 선택해주세요!')
                        isWrongIndex = True
                
                elif choosen_product == 0:
                    print('구매 스킵')
                    isWrongIndex = False
                
                else:
                    print('올바른 상점 물품 번호를 입력해주세요!')
                    isWrongIndex == True
                    
                
    def dice_skill(self, player_arr, player_info, cur_dice):
        print("{0} - 주사위 스킬이 발동됩니다!".format(player_info.player))

        # 파이썬에는 switch가 없으므로 if로 구현해야함!
        if cur_dice == 2:
            print("모두의 자금에서 50만원을 깐다.")
            
            for i in range(len(player_arr)):
                player_arr[i].withdraw(500000)
        elif cur_dice == 4:
            
            player_info.money *= (player_info.money * 0.95)
            print("자신의 자금의 5%가 증발한다. 남은돈:", player_info.money)
            # TODO: 숙제
        elif cur_dice == 6:
            print("모두의 자금에 30만원을 추가한다.")
            
            for i in range(len(player_arr)):
                player_arr[i].deposit(300000)
        
    def start_game(self):
        self.player = int(input("몇 명이 플레이 하는 게임입니꽈 ? "))
        
        for i in range(self.player):
            name = input("플레이어 이름을 적어주세요")
            self.player_arr.append(Player(name)) #객체 배열 생성
            
            print('플레이어', self.player_arr[i].player, '생성!')
            
        while (not(self.game_end)):
            
            for i in range(self.player):
                
                for j in range(self.player):
                    if self.player_arr[j].add_times > 0:
                        self.loop_on = True
                
                if self.loop_on:
                    if 0 == self.player_arr[i].add_times:
                        continue
                
                if self.loop_player_no != -1:
                    if i != self.loop_player_no:
                        continue
                
                if self.player_arr[i].add_times > 0:
                    self.player_arr[i].add_times -= 1
                    print(self.player_arr[i].player, '플레이어가', self.player_arr[i].add_times, '회 연속 굴리기 남았습니다.')
                    
                    self.loop_player_no = i
                    
                    if self.player_arr[i].add_times == 0:
                        print(self.player_arr[i].player, '플레이어의 연속 굴리기가 끝났습니다')
                        self.loop_player_no = -1
                        self.loop_on = False
                    
                if self.player_arr[i].tied_times > 0:
                    print(self.player_arr[i].player, '플레이어가', self.player_arr[i].tied_times, '회 정지된 상태이므로 건너뜁니다.')
                    self.player_arr[i].tied_times -= 1
                    continue
                
                self.player_arr[i].run_dice()
                cur_dice = self.player_arr[i].get_dice()
                self.dice_sum += cur_dice
                
                if (not(cur_dice % 2)):
                    self.dice_skill(self.player_arr, self.player_arr[i], cur_dice)
                    
                self.player_arr[i].update_location(cur_dice)
                    
                if (self.player_arr[i].get_lap() == 3):
                    self.game_end = True
                    break
                    
            if (self.game_end):
                break
                
            if (not(self.dice_sum % 2)):
                self.open_secret_store()
                
            self.dice_sum = 0
            
        # 정산을 통해 승자 판정(게임을 먼저 끝내도 4번만 나왔으면 질 수도 있음)
        # TODO: 숙제
        
        winner_money = 0
        winner_name = ''
        
        for i in range(len(self.player_arr)):
            print(self.player_arr[i].__dict__)
            
            if winner_money < self.player_arr[i].money:
                winner_money = self.player_arr[i].money
                winner_name = self.player_arr[i].player
                
        print('최종 승자는', winner_money, '를 보유한', winner_name, '플레이어입니다!')
        
class Player(Dice):
    def __init__(self, player):
        super(Player, self).__init__(player)
        # 몇 바퀴 돌았는지
        self.lap = 0
        # 현재 위치
        self.location = 0
        # 소지금
        self.money = 0
        self.tied_times = 0
        self.add_times = 0
        
        print("Player 생성자 호출!")
        
    def update_location(self, dice):
        self.location += dice
        
        if (self.location >= 40):
            self.location -= 40
            self.lap += 1
            
            self.money += 1000000
            
        print("위치값 갱신 lap - {0}, location - {1}".format(self.lap, self.location))
            
    def get_lap(self):
        return self.lap
        
    def withdraw(self, money):
        self.money -= money
        print("출금: 내돈 = {0}".format(self.money))
        
    def deposit(self, money):
        self.money += money
        print("입금: 내돈 = {0}".format(self.money))
        
        
board_game = Game()
board_game.start_game()

print(board_game.__dict__)
```
