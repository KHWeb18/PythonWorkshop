```python
# 3. 보드게임 만들기

# 한 바퀴를 도는데 총 주사위 눈금이 40칸 필요하다.
# 1) 한 바퀴를 돌면 자금으로 100만원이 주어진다.
# 2) 주사위의 특수 능력을 활용할 수 있다.
#    2-1) 숫자 2: 모두의 자금에서 50만원 빼기
#    2-2) 숫자 4: 자신의 자금의 5% 빼기
#    2-3) 숫자 6: 모두의 자금에 30만원을 추가
#    2-4) 모든 플레이어들이 주사위를 돌린 눈금이 짝수일 경우 비밀 상점이 열린다.
#         비밀 상점 물품: 레드 티켓(지정한 상대의 주사위가 5턴 동안 정지함) - 100만원
#                        옐로 티켓(지정한 상대의 주사위가 2턴간 정지함) - 50만원
#                        카오스 티켓(아무나 한 명 2턴간 정지함) - 10만원
#                        뽑기 티켓(1만원, 2만원, 5만원, 10만원, 20만원, 50만원, 100만원) - 10만원
#                        연속 굴리기(1회, 2회, 3회, 4회) - 20만원
#    2-5) 플레이어가 필드를 3바퀴 돌면 게임이 종료된다. (주사위 눈금 총 120칸 필요)
#    2-6) 게임이 종료된 이후 가장 많은 자금을 가지고 있는 사람이 경기에 승리하게 된다.
# 게임 팁: 1) 일단 돈을 많이 모은다.
#          2) 비밀 상점을 활용하여 대박(뽑기)을 노리거나 상대방을 행동 불능에 빠지게 한다.
#          3) 유리한 상황에서 연속 굴리기를 통해 게임을 빠르게 끝내도록 유도한다.
#          4) 다수가 함께 플레이 할 경우 연속 굴리기를 통해 정치도 가능하다 ?
```

```python
from random import *

class Dice:
    def __init__(self, player):
        self.player = player
        
    def run_dice(self):
        self.dice = randint(1, 6)
        print('{0}의 주사위: {1}'.format(self.player, self.dice))
    
    def who_win(self, versus):
        if (self.dice > versus.dice):
            print('{0} 승리 ({1} vs {2})'.format(self.player, self.dice, versus.dice))
        elif (self.dice < versus.dice):
            print('{0} 승리 ({1} vs {2})'.format(versus.player, self.dice, versus.dice))
        else:
            print('무승부 ({0} vs {1})'.format(self.dice, versus.dice))
            
    def get_dice(self):
        return self.dice
            
player1 = Dice('플레이어1')
player1.run_dice()

player2 = Dice('플레이어2')
player2.run_dice()

player1.who_win(player2)
```

```python
class Game:
    def __init__(self):
        self.secret_store = False
        self.game_end = False
        self.use_chaos_ticket = False
        self.player_arr = []
        self.dice_sum = 0
        
    def open_secret_store(self, player_arr):
        print('비밀상점 오픈')
   
        player_info = {}
        player_length = len(player_arr)
        
        for i in range(player_length):
            player_core_info = []
            player_core_info.append(self.player_arr[i].get_lap())
            player_core_info.append(self.player_arr[i].get_location())
            player_core_info.append(self.player_arr[i].get_money())
            
            player_info[self.player_arr[i].get_player()] = player_core_info
        
        money_gacha = [10000, 20000, 50000, 100000, 200000, 500000, 1000000]
        
        print("player_length = ", player_length)
        
        for i in range(player_length):
            while True:
                not_finish = True
                
                print('현재 유저 리스트: ')

                for j in range(player_length):
                    print(player_arr[j].__dict__)
                
                print('1. 레드티켓(5회 정지), 2. 옐로티켓(2회 정지), 3. 카오스티켓(랜덤 2회 정지), 4. 뽑기(돈), 5. 연속굴리기(1 ~ 4회), 6. 넘어가기')
                
                select = int(input('선택한 아이템 번호를 입력하세요. '))

                if (select == 1):
                    print('레드티켓을 선택하셨습니다.')

                    while not_finish:
                        idx = int(input('적용할 상대를 선택하세요. '))
                        if (idx < player_length and idx >= 0):
                            player_arr[idx].add_stop_count(5)
                            not_finish = False
                            
                    break
                elif (select == 2):
                    print('옐로티켓을 선택하셨습니다. ')
                    
                    while not_finish:
                        idx = int(input('적용할 상대를 선택하세요. '))
                        if (idx < player_length and idx >= 0):
                            player_arr[idx].add_stop_count(2)
                            not_finish = False
                            
                    break
                elif (select == 3):
                    print('카오스티켓을 선택하셨습니다.')
                    
                    while not_finish:
                        idx = randint(0, player_length)
                        if (idx < player_length and idx >= 0):
                            player_arr[idx].add_stop_count(2)
                            not_finish = False
                    
                    break
                elif (select == 4):
                    print('뽑기를 선택하셨습니다. 뽑기를 시작합니다. ')
                    
                    gacha_idx = randint(0, len(money_gacha) - 1)
                    print('가챠 인덱스 = ', gacha_idx)
                    
                    player_arr[i].add_money(money_gacha[gacha_idx])
                    
                    break
                elif (select == 5):
                    print('연속굴리기를 선택하셨습니다. ')
                    
                    rand_count = randint(1, 4)
                    print("rand_count = ", rand_count)
                    
                    player_arr[i].set_repeat_count(rand_count)
                    
                    break
                elif (select == 6):
                    print('아무것도 선택하지 않으므로 넘어갑니다.')
                    break
                else:
                    print('다시 선택해 주세요.')
            
        print('비밀상점 종료')
        
    def dice_skill(self, player_arr, player_info, cur_dice):
        print('{0}의 주사위 스킬'.format(player_info.player))

        if cur_dice == 2:
            print('플레이어 모두 50만원 감소')
            
            for i in range(len(player_arr)):
                player_arr[i].withdraw(500000)
        elif cur_dice == 4:
            print('본인의 자금 5% 감소')

            player_info.withdraw_percent(5)
        elif cur_dice == 6:
            print('플레이어 모두 30만원 추가')
            
            for i in range(len(player_arr)):
                player_arr[i].deposit(300000)
        
        
    def start_game(self):
        self.player = int(input('플레이어 수 '))
        
        for i in range(self.player):
            name = input('플레이어 이름')
            self.player_arr.append(Player(name))
            self.player_arr[i].repeat_count = 0
            
        while (not(self.game_end)):
            for i in range(self.player):

                repeat_count = self.player_arr[i].repeat_count
                stop_count = self.player_arr[i].stop_count
                
                tmp_count = repeat_count - stop_count
                
                if (repeat_count == stop_count):
                    self.player_arr[i].repeat_count = 0
                    self.player_arr[i].stop_count = 0
                
                
                while tmp_count >= 0:
                    if (self.player_arr[i].stop_count == 0 or tmp_count == 0):
                        self.player_arr[i].run_dice()
                        cur_dice = self.player_arr[i].get_dice()
                        self.dice_sum += cur_dice

                        if (not(cur_dice % 2)):
                            self.dice_skill(self.player_arr, self.player_arr[i], cur_dice)

                        self.player_arr[i].update_location(cur_dice)

                        if (self.player_arr[i].get_lap() == 3):
                            self.game_end = True
                            break
                            
                        tmp_count -= 1
                        
                if (self.player_arr[i].stop_count != 0):
                    self.player_arr[i].stop_count -= 1
                    
                self.player_arr[i].repeat_count = 0
                    
            if (self.game_end):
                break
                
            if (not(self.dice_sum % 2)):
                self.open_secret_store(self.player_arr)
                
            self.dice_sum = 0

        income = {}
        
        for i in range(len(self.player_arr)):
            print(self.player_arr[i].__dict__)
            
            income[self.player_arr[i].get_player()] = self.player_arr[i].get_money()
            
        print("income: ", income)
        
        sorted_income = sorted(income.items(), key=value_support, reverse=True)
        
        print("income: ", sorted_income)
        
        print("최종 승자: {0}".format(sorted_income[0]))
        
class Player(Dice):
    def __init__(self, player):
        super(Player, self).__init__(player)
        # 몇 바퀴 돌았는지
        self.lap = 0
        # 현재 위치
        self.location = 0
        # 소지금
        self.money = 0
        # 정지 횟수
        self.stop_count = 0
        # 연속 횟수
        self.repeat_count = 0
        
    def update_location(self, dice):
        self.location += dice
        
        if (self.location >= 40):
            self.location -= 40
            self.lap += 1
            
            self.money += 1000000
            
    def get_lap(self):
        return self.lap
    
    def get_location(self):
        return self.location
        
    def withdraw(self, money):
        self.money -= money
        print('내 소지금: {0}'.format(self.money))
        
    def deposit(self, money):
        self.money += money
        print('내 소지금: {0}'.format(self.money))
        
    def withdraw_percent(self, percent):
        #self.money = self.money - self.money * 5 / 100
        #self.money = self.money * (1 - 5 / 100)
        #self.money -= self.money * (5 / 100)
        # a x b = ab
        self.money -= self.money * percent / 100
        print('내 소지금: {0}'.format(self.money))
        
    def get_money(self):
        return self.money
    
    def get_player(self):
        return self.player
    
    def add_stop_count(self, count):
        self.stop_count += count
        
    def add_money(self, money):
        self.money += money
        
    def set_repeat_count(self, count):
        self.repeat_count = count
        
board_game = Game()
board_game.start_game()

print(board_game.__dict__)
```
